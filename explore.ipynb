```python
import pandas as pd
```

```python
data = pd.read_csv("data/Train_details_22122017.csv")
```

```python
data
```

```python
# Melt the dataframe to have a single time column and an indicator column
columns = list(data.columns)
# columns = [x for x in columns if x not in ["Source Station",	"Source Station Name",	"Destination Station",	"Destination Station Name"]]
print(columns)
arrival_columns = columns.copy()
arrival_columns.remove("Departure Time")
departure_columns = columns.copy()
departure_columns.remove("Arrival time")
arrival_df = data[arrival_columns].copy()
arrival_df.rename(columns={'Arrival time': 'Time'}, inplace=True)
arrival_df['Type'] = 'Arrival'

departure_df = data[departure_columns].copy()
departure_df.rename(columns={'Departure Time': 'Time'}, inplace=True)
departure_df['Type'] = 'Departure'
```

```python
arrival_df
```

```python
combined_df = pd.concat([arrival_df, departure_df]).sort_values(by=['Train No', 'SEQ', 'Type']).reset_index(drop=True)
```

```python
combined_df
```

```python
combined_df.to_csv("data/data_reorged.csv", index=False)
```
